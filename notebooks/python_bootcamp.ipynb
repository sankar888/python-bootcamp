## Python Syntax
- Python was designed for readability, and has some similarities to the English language with influence from mathematics.
- Python uses new lines to complete a command, as opposed to other programming languages which often use semicolons or parentheses.
- Python relies on indentation, using whitespace, to define scope; such as the scope of loops, functions and classes. Other programming languages often use curly-brackets for this purpose.

````python
# Python Syntax - semicolon
# python uses new lines to complete a command
a = 1 + 2
print(a)

# Python does not require semi-colons to terminate statements. but it is also not illegal to use semi-colon. 
# A semicolon in Python denotes separation, rather than termination. It allows you to write multiple statements on the same line.
a = 2 + 3; print(a)
````

## Variables
Variables are containers for storing data values. Variables do not need to be declared with any particular type, and can even change type after they have been set. A variable is created the moment you first assign a value to it. Variables are case sensitive (python is case sensitive). 

Rules for naming a variable
- A variable can have a short name (like x and y) or a more descriptive name (age, carname, total_volume). Rules for Python variables:
- A variable name must start with a letter or the underscore character
- A variable name cannot start with a number
- A variable name can only contain alpha-numeric characters and underscores (A-z, 0-9, and _ )
- Variable names are case-sensitive (age, Age and AGE are three different variables)
- A variable name cannot be any of the Python keywords.

### Strings
- Python strings can be single qouted or double quotted
- triple quotes string preserve whitespace, tabs and new lines
- the escape character in string is \
- the newline (\n), tab (\t) are interpreted in strings
- type of string is str
- objects of other types can be converted to string with str() function

````python
# PYTHON BUIL-IN DATA TYPES

# Python strings can be single qouted or double quotted
double_quoted_string = "friday"
single_quoted_string = 'friday'
print(type(double_quoted_string), type(single_quoted_string))

# triple quoted string preserve white space, tabs and new line
triple_quoted_string = """
    monday
    tuesday
    wednesday
"""
triple_quoted_string1 = '''
monday
tuesday
    wednesday
thursday
'''
print(triple_quoted_string, triple_quoted_string1)

# the newline (\n), tab (\t) are interpreted in strings
print("hai\thow\tare\tu?")

# objects of other types can be converted to string with str() function
a = 10
b = str(a)
print(b, "type of b is ", type(b)) 
````

### Numeric Type and Boolean
Numeric types of python are int, float and complext types. vartiables could not be declared in python. types are infered during runtime. we could fix the type of an object using casting. we could do type casting using the below functions or built-ins
- int()
- float()
- complex()

Boolean is represented by built-in objects ```True``` and ```False```. bool is  subclass of int. True is equal to number 1 and false is equal to number 0. built-in function bool() is used to convert any value to truth or false value.

````python
# Numeric types in python
# int
a = 10
print("type of 10 is ", type(a))

# float
a = 10.3
print("type of 10.3 is ", type(a))

# complex types
a = 10.3j
print("type of 10.3j is ", type(a))

# type casting - can be done using
a = 10 #type int
a = float(a)
print("type of float(10) is", type(a))
print("value of int(10.3) is", int(10.3), "type of int(10.3) is", type(int(10.3)))
print("value of complex(10.3) is", complex(10.3), "type of complex(10.3) is", type(complex(10.3)))

# Boolean type demo
# Boolean is represented by built-in objects ```True``` and ```False```. 
# It is implemented a integer. True is equal to number 1 and false is equal to number 0
a = True
print("value of True is", True, "type of True is", type(True))
print("True == 1", True == 1)
print("False == 0", False == 0)
print("value of float(True) should be 1.0:", float(True))
print("value of bool('hi') should be True:", bool('hi'))
print("value of bool("") should be False as empty string denote a false value:", bool(""))
````

### None Type
None Type is an object type which denotes nothing (null). the truth value of None is False 

- functions which return nothing is returning None
- None can be used to check if the value is empty
- use is and is not opeator when checking for 'None' as it is an object.

````python
# None value is a special builtin value which represents Value of type NoneType
# None is used to represent null or nothing
a = None
print("value of a is", a, "type of a is", type(a))

# function which doesn't return anything 
print(print("print() function doesnot return anything"))

# truth value of None is False
print("truth value of None is False:", bool(None), ",negation of None is True", not None)
````
